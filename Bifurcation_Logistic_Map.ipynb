```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def scatter(x, y, title_text, x_labeltext, y_labeltext, title_size):
    fig = plt.figure(figsize = (25,25))
    plt.title(title_text, fontsize = title_size)
    ax = fig.add_subplot(111)
    ax.scatter(x,y, alpha = 0.5, c = 'k', marker = ".")
    ax.set(ylabel = y_labeltext, xlabel = x_labeltext)
    ax.legend(loc = 'best')
    
```

```python
# parameters
# mu, x_0
bid = 500
d = {'mu': [], 'x': []}

for mu in np.linspace(2.4, 4, num = bid ):
    x = 0.5
    for n in range(1, 5000):
        x = mu * x * (1-x)
    
    for i in range(1, 1000):
        x = mu * x * (1-x)
        d['mu'].append(mu)
        d['x'].append(x)
        
df = pd.DataFrame(data = d)
```

```python
scatter(d['mu'], d['x'], 'Bifurcation of Logistic Map', 'mu', '$x_*$', title_size = 50)
```
